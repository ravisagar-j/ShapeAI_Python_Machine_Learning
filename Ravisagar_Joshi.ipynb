Adding new column to the dataframe, Named as MEDV

```python
boston['MEDV'] = df.target
boston.head()
```

Check if the dataset contains any null values or not?

```python
from sklearn.model_selection import train_test_split

X = boston.drop('MEDV', axis = 1)
Y = boston['MEDV']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.5, random_state = 5)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

import the Linear Regression model from sklearn and train the Dataset

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

Create a model using Linear Regrassion

```python
lin_model = LinearRegression()

lin_model.fit(X_train, Y_train)
```

```python
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))

print("the model performance for training set")
print('RMSE is {}'.format(rmse))
print('\n')

#Testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))

print("the model performance for training set")
print('RMSE is {}'.format(rmse))
```
